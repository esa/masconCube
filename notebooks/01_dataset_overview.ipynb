```python
# Ensure that changes in imported module (gravann most importantly) are autoreloaded
%load_ext autoreload
%autoreload 2

from matplotlib import pyplot as plt
from mascon_cube.constants import GROUND_TRUTH_DIR
import numpy as np
from mascon_cube.data.mesh import get_mesh
```

```python
asteroids = [f.name for f in sorted(GROUND_TRUTH_DIR.iterdir()) if f.is_dir()]
```

```python
base_dim = 3.3
nrows = 2
ncols = len(asteroids) // nrows if len(asteroids) % nrows == 0 else len(asteroids) // nrows + 1
figsize = (base_dim*ncols, base_dim*nrows)
fig, axs = plt.subplots(nrows, ncols, squeeze=True, figsize=figsize)

for asteroid, ax in zip(asteroids, axs.flatten()):
    img = plt.imread(GROUND_TRUTH_DIR / asteroid / "combined_plot.png")
    ax.imshow(img)
    ax.axis("off")
    ax.set_title(asteroid)
fig.tight_layout(pad=1)
plt.show()
    
```

```python
for asteroid in asteroids:
    print(asteroid + ":")
    points, triangles = get_mesh(asteroid)
    print(f"\tpoints: {points.shape[0]}")
    print(f"\ttriangles: {triangles.shape[0]}")

    
```
